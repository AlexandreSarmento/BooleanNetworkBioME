```python
def dNodesdt(O2,C0,params,dt,niter):
    
    ss,jO2PHD,jO2FIH,p300,ksHIF1ac,kinHIF1ac,kHIF1aoi,koutHIF1an,kdeHIF1ac,jdeHIF1ac,kd1HIF1ac,jd1HIF1ac,kd2HIF1ac,jd2HIF1ac,jd2HIF1aac,jdp53c,dHIF1ac,kd1HIF1aac,jd1HIF1aac,kd2HIF1aac,dHIF1aac,kinHIF1aac,kHIF1aaoi,koutHIF1aan,kd2HIF1aan,jd2HIF1aan,jd2HIF1an,jdp53n,jdp53nP,dHIF1aan,kd2HIF1an,dHIF1an,kacHIF1an,jacHIF1an,jacp53nP,kdeHIF1anAC,jdeHIF1anAC,dHIF1anAC,kacFIH0,kacFIH,jacFIH,kdeFIH,jdeFIH,FIHtot,ksPHD_tot0,ksPHD_tot1,jksPHD_tot1,ksPHD_tot2,jksPHD_tot2,ksPHD_tot3,jksPHD_tot3,dPHD_tot,kacPHD0,kacPHD,kdePHD,ksPNUTS0,ksPNUTS,jsPNUTS,dPNUTS,kacATR0,kacATR,KO2,jacATR,kdeATR,jdeATR,ATRtot,ksp53c,kdp53c,kinp53c,kp53oi,koutp53n,dp53c,kacp53n1,kacp53n2,jacp53n,kdep53nP,jdep53nP,kdp53n,dp53n,kacp53nP,kdep53nAC,jdep53nAC,dp53nP,kdp53nP,dp53nAC,ksMdm2c0,ksMdm2c,jsMdm2c,kinMdm2c,sita,jinMdm2c,koutMdm2n,joutMdm2n,kdMdm2c,dMdm2n,ksSHIP0, ksSHIP,jsSHIP,dSHIP,kp2,jp2,kp3,jp3,PIPtot,kacAKT,jacAKT,kdeAKT,jdeAKT,AKTtot,ksPFKL0,ksPFKL1,jsPFKL1,ksPFKL2,jsPFKL2,dPFKL,ksVEGF0,ksVEGF,jsVEGF,dVEGF,ksBax0,ksBax,jsBax,dBax= params
    
    koutHIF1an = kHIF1aoi*kinHIF1ac
    koutHIF1aan = kHIF1aaoi*kinHIF1aac
    koutp53n = kp53oi*kinp53c
    koutMdm2n = kinMdm2c*sita
    
    # ,FIH,PHD,ATR,PIP2,AKT
    # np.zeros(len(t)), np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),
    HIF1ac,HIF1aac,HIF1aan,HIF1an,HIF1anAC,FIHa,PHDtot,PHDa,PNUTS,ATRa,p53c,p53n,p53nP,p53nAC,Mdm2c,Mdm2n,SHIP,PIP3,AKTa,PFKL,VEGF,Bax,p53tot,HIF1atot,Mdm2tot,FIH,PHD,ATR,PIP2,AKT = np.zeros(len(t)), np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)), np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t)),np.zeros(len(t))
    
    
    HIF1ac[0],HIF1aac[0],HIF1aan[0],HIF1an[0],HIF1anAC[0],FIHa[0],PHDtot[0],PHDa[0],PNUTS[0],ATRa[0],p53c[0],p53n[0],p53nP[0],p53nAC[0],Mdm2c[0],Mdm2n[0],SHIP[0],PIP3[0],AKTa[0],PFKL[0],VEGF[0],Bax[0] = 0.14086,0.85963,0.79719,0.10293,0.10729,1.99239,7.69505,3.65788,0.001,0.01019,0.101,0.36969,0.01904,0.0053,0.13963,1.66037,0.02857,0.99824,0.94319,0.04076,0.02007,0.05
    
    for i in range(1,niter):
        
        HIF1ac[i] = HIF1ac[i-1] + (ksHIF1ac-kinHIF1ac*HIF1ac[i-1]+koutHIF1an*HIF1an[i-1] - dHIF1ac*HIF1ac[i-1] - kdeHIF1ac*FIHa[i-1]*HIF1ac[i-1]/(HIF1ac[i-1]+jdeHIF1ac)- kd1HIF1ac*PHDa[i-1]*HIF1ac[i-1]/(HIF1ac[i-1]+jd1HIF1ac)- kd2HIF1ac*Mdm2c[i-1]*HIF1ac[i-1]/jd2HIF1ac/(1+HIF1ac[i-1]/jd2HIF1ac+HIF1aac[i-1]/jd2HIF1aac+p53c[i-1]/jdp53c))*dt
         
        HIF1aac[i] = HIF1aac[i-1] + (kdeHIF1ac*FIHa[i-1]*HIF1ac[i-1]/(HIF1ac[i-1]+jdeHIF1ac)- dHIF1aac*HIF1aac[i-1] - kd1HIF1aac*PHDa[i-1]*HIF1aac[i-1]/(HIF1aac[i-1]+jd1HIF1aac) - kd2HIF1aac*Mdm2c[i-1]*HIF1aac[i-1]/jd2HIF1aac/(1+HIF1ac[i-1]/jd2HIF1ac+HIF1aac[i-1]/jd2HIF1aac+p53c[i-1]/jdp53c) - kinHIF1aac*HIF1aac[i-1]+koutHIF1aan*HIF1aan[i-1])*dt
        
        HIF1aan[i] = HIF1aan[i-1] + (kinHIF1aac*HIF1aac[i-1]-koutHIF1aan*HIF1aan[i-1]-dHIF1aan*HIF1aan[i-1] - kd2HIF1aan*Mdm2n[i-1]*HIF1aan[i-1]/jd2HIF1aan/(1+HIF1aan[i-1]/jd2HIF1aan+HIF1an[i-1]/jd2HIF1an+p53n[i-1]/jdp53n+p53nP[i-1]/jdp53nP))*dt
       
        HIF1an[i] = HIF1an[i] + (kinHIF1ac*HIF1ac[i-1]-koutHIF1an*HIF1an[i-1] - dHIF1an*HIF1an[i-1] - kd2HIF1an*Mdm2n[i-1]*HIF1an[i-1]/jd2HIF1an/(1+HIF1aan[i-1]/jd2HIF1aan+HIF1an[i-1]/jd2HIF1an+p53n[i-1]/jdp53n+p53nP[i-1]/jdp53nP)- kacHIF1an*p300*HIF1an[i-1]/jacHIF1an/(1+HIF1an[i-1]/jacHIF1an+p53nP[i-1]/jacp53nP) + kdeHIF1anAC*HIF1anAC[i-1]/(HIF1anAC[i-1]+jdeHIF1anAC))*dt
       
        HIF1anAC[i] = HIF1anAC[i-1] + (kacHIF1an*p300*HIF1an[i-1]/jacHIF1an/(1+HIF1an[i-1]/jacHIF1an+p53nP[i-1]/jacp53nP)-kdeHIF1anAC*HIF1anAC[i-1]/(HIF1anAC[i-1]+jdeHIF1anAC)-dHIF1anAC*HIF1anAC[i-1])*dt
      
        HIF1atot[i] = HIF1ac[i-1] + HIF1aac[i-1] + HIF1aan[i-1] + HIF1an[i-1] + HIF1anAC[i-1]
       
        FIH[i-1] = FIHtot - FIHa[i-1]
        
        FIHa[i] = FIHa[i-1] + (kacFIH0+kacFIH*O2*ss/(O2*ss+jO2FIH)*FIH[i-1]/(FIH[i-1]+jacFIH)-kdeFIH*FIHa[i-1]/(FIHa[i-1]+jdeFIH))*dt
        
        PHDtot[i] = PHDtot[i-1] + (ksPHD_tot0+ksPHD_tot1*HIF1aan[i-1]**4/(HIF1aan[i-1]**4+jksPHD_tot1**4)+ ksPHD_tot2*HIF1anAC[i-1]**4/(HIF1anAC[i-1]**4+jksPHD_tot2**4)+ksPHD_tot3*HIF1an[i-1]**4/(HIF1an[i-1]**4+jksPHD_tot3**4)- dPHD_tot*PHDtot[i-1])*dt
        
        PHD = PHDtot[i-1] - PHDa[i-1]
        # dPHDa/dt = kacPHD0+kacPHD*O2*ss/(O2*ss+jO2PHD)*PHD-kdePHD*PHDa
        PHDa[i] = PHDa[i-1] + (kacPHD0+kacPHD*O2*ss/(O2*ss+jO2PHD)*PHD-kdePHD*PHDa[i-1])*dt
        
        PNUTS[i] = PNUTS[i-1] + (ksPNUTS0+ksPNUTS*HIF1anAC[i-1]**4/(HIF1anAC[i-1]**4+jsPNUTS**4)-dPNUTS*PNUTS[i-1])*dt
        
        ATR = ATRtot - ATRa[i-1]
       
        ATRa[i] = ATRa[i-1] + ((kacATR0+kacATR*KO2/(O2*ss+KO2)*ATRa[i-1])*ATR/(ATR+jacATR)-kdeATR*ATRa[i-1]/(ATRa[i-1]+jdeATR))*dt
            
        p53c[i] = p53c[i-1] + (ksp53c-kinp53c*p53c[i-1] + (koutp53n*p53n[i-1]-dp53c*p53c[i-1] - kdp53c*Mdm2c[i-1]*p53c[i-1]/jdp53c/(1+HIF1ac[i-1]/jd2HIF1ac+HIF1aac[i-1]/jd2HIF1aac+p53c[i-1]/jdp53c)))*dt
        
        p53n[i] = p53n[i-1] + (kinp53c*p53c[i-1]-koutp53n*p53n[i-1] - dp53n*p53n[i-1] - (kacp53n1*ATRa[i-1]+kacp53n2*PNUTS[i-1])*p53n[i-1]/(p53n[i-1]+jacp53n)+kdep53nP*p53nP[i-1]/(p53nP[i-1]+jdep53nP)- kdp53n*Mdm2n[i-1]*p53n[i-1]/jdp53n/(1+HIF1aan[i-1]/jd2HIF1aan+HIF1an[i-1]/jd2HIF1an+p53n[i-1]/jdp53n+p53nP[i-1]/jdp53nP))*dt
     
        p53nP[i] = p53nP[i-1] + ((kacp53n1*ATRa[i-1]+kacp53n2*PNUTS[i-1])*p53n[i-1]/(p53n[i-1]+jacp53n)-kdep53nP*p53nP[i-1]/(p53nP[i-1]+jdep53nP)-kdp53nP*Mdm2n[i-1]*p53nP[i-1]/jdp53nP/(1+HIF1aan[i-1]/jd2HIF1aan+HIF1an[i-1]/jd2HIF1an+p53n[i-1]/jdp53n+p53nP[i-1]/jdp53nP)-kacp53nP*p300*p53nP[i-1]/jacp53nP/(1+HIF1an[i-1]/jacHIF1an+p53nP[i-1]/jacp53nP)+kdep53nAC*p53nAC[i-1]/(p53nAC[i-1]+jdep53nAC)-dp53nP*p53nP[i-1])*dt
       
        p53nAC[i] = p53nAC[i-1] + (kacp53nP*p300*p53nP[i-1]/jacp53nP/(1+HIF1an[i-1]/jacHIF1an+p53nP[i-1]/jacp53nP)-kdep53nAC*p53nAC[i-1]/(p53nAC[i-1]+jdep53nAC) - dp53nAC*p53nAC[i-1])*dt
       
        p53tot[i] = p53c[i-1] + p53n[i-1] + p53nP[i-1] + p53nAC[i-1]
       
        Mdm2c[i] = Mdm2c[i-1] + (ksMdm2c0+ksMdm2c*p53nAC[i-1]**4/(p53nAC[i-1]**4+jsMdm2c**4)- kdMdm2c*Mdm2c[i-1] - kinMdm2c*AKTa[i-1]*Mdm2c[i-1]/(Mdm2c[i-1]+jinMdm2c)+koutMdm2n*Mdm2n[i-1]/(Mdm2n[i-1]+joutMdm2n))*dt

        Mdm2n[i] = Mdm2n[i-1] + (kinMdm2c*AKTa[i-1]*Mdm2c[i-1]/(Mdm2c[i-1]+jinMdm2c)- koutMdm2n*Mdm2n[i-1]/(Mdm2n[i-1]+joutMdm2n)-dMdm2n*Mdm2n[i-1])*dt

        Mdm2tot[i] = Mdm2c[i-1] + Mdm2n[i-1]

        SHIP[i] = SHIP[i-1] + (ksSHIP0+ksSHIP*p53nAC[i-1]**4/(p53nAC[i-1]**4+jsSHIP**4)-dSHIP*SHIP[i-1])*dt
        
        PIP2 = PIPtot-PIP3[i-1]

        PIP3[i] = PIP3[i-1] + (kp2*PIP2/(PIP2+jp2)-kp3*SHIP[i-1]*PIP3[i-1]/(PIP3[i-1]+jp3))*dt
       
        AKT = AKTtot - AKTa[i-1]

        AKTa[i] = AKTa[i-1] + (kacAKT*PIP3[i-1]*AKT/(AKT+jacAKT)-kdeAKT*AKTa[i-1]/(AKTa[i-1]+jdeAKT))*dt

        PFKL[i] = PFKL[i-1] + (ksPFKL0+ksPFKL1*HIF1an[i-1]**4/(HIF1an[i-1]**4+jsPFKL1**4)+ksPFKL2*HIF1aan[i-1]**4/(HIF1aan[i-1]**4+jsPFKL2**4)-dPFKL*PFKL[i-1])*dt

        VEGF[i] = VEGF[i-1] + (ksVEGF0+ksVEGF*HIF1anAC[i-1]**4/(HIF1anAC[i-1]**4+jsVEGF**4)-dVEGF*VEGF[i-1])*dt

        Bax[i] = Bax[i-1] + (ksBax0+ksBax*p53nAC[i-1]**4/(p53nAC[i-1]**4+jsBax**4)-dBax*Bax[i-1])*dt
    
    
    return [HIF1ac,HIF1aac,HIF1aan,HIF1an,HIF1anAC,HIF1atot,FIHa,PHDtot,PHDa,PNUTS,ATRa,p53c,p53n,p53nP,p53nAC,p53tot,Mdm2c,Mdm2n,Mdm2tot,SHIP,PIP3,AKTa,PFKL,VEGF,Bax]
    
```
